### Generación Aleatoria de Números

```python
import numpy as np
```

```python
# Funcion random que genera un entero (int) entre 1 y 100
np.random.randint(1,100)
```

```python
# La forma mas clasica qde generar un número aleatorio es entre 0 y 1 (con decimales)
np.random.random()
```

```python
## Función que genera una lista de n numeros aleatorios enteros dentro del intervalo [a,b].
# Range indica la cantidad de numeros a generar
# Append se usa para meter numeros al final del arreglo 
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x    
```

```python
randint_list(25, 1, 50)
```

```python
# Para evitar hacer una función, ya existe una
import random
```

```python
# Acá hacemos una función que genere numeros de 0 a 100, pero que sean multiplos de 7
random.randrange(0, 100, 7)
```

```python
# En esta funcion, generamos 10 numeros aleatorios de 0 a 100, pero multiplos de 7 + 1 (Es el 1 en (1, 100, 7))
for i in range(10):
    print(random.randrange(1, 100, 7))
```
