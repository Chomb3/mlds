# Agregación de Datos por Categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

# Generaremos 500 datos aleatorios para gender e income
for i in range(0, 500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
# Creación de un dataframe con datos aleatorios para peso, altura, edad e ingresos
## z -> N(0,1)
## N(m,s) -> m + s * z

height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.rand(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.head()
```

## Agrupación de Datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
## Para filtrar solo un grupo
grouped_gender.get_group("Female")
```

```python
## Filtrar por mas de una categoria
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

#### Agregación de Datos

```python
double_group.aggregate(
    {
        "Income" : np.sum,
        "Age" : np.mean,
        "Height" : np.std
    }
)
```

```python
## Si se quiere la tipificacion (funcion lambda) de un dato -> promedio/desviacion estandar
## Tipificar es realizar una transformacion lineal para que la nueva variable tenga promedio 0 y desv. estandar 1
double_group.aggregate(
    {
        "Age" : np.mean,
        "Height": lambda h:(np.mean(h))/np.std(h)
    }
)
```

```python
## Si queremos aplicarle funciones a todas las columnas
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
## Para obtener lambda (tipificacion) de cada columna
double_group.aggregate([lambda x: np.mean(x)/np.std(x)])
```

## Filtrado de Datos

```python
# Por ejemplo filtrar grupos cuya suma de edades sea mayor a 2400
double_group["Age"].filter(lambda x: x.sum()>2400)
```

## Transformación de Variables

```python
zscore = lambda x: (x - x.mean()) / x.std()
```

```python
z_group = double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Age"])
```

```python
#Reemplazar los na con el promedio (Aunque en este dataframe no habia na)
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```

## Dividir usando una Funcion Shuffle

```python
import sklearn
```

```python
data.head()
```

```python
## Desordenamos los datos
data = sklearn.utils.shuffle(data)
```

```python
cut_id = int(0.75 * len(data))

## train_data serán los datos desde el inicio hasta cut_id (75% de los datos) -> [ :75%]
train_data = data[:cut_id]

## test_data serán los datos desde el 75% + 1 dato, hasta el final -> [75% + 1 : ]
test_data = data[cut_id + 1:]
```
