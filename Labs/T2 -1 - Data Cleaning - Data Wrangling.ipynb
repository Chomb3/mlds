##### Eliminando las Columnas Innecesarias

```python
# En dedired_columns dejamos las columnas que queremos eliminar
desired_columns = ["Account Length",  "VMail Message", "Day Calls"]
desired_columns
```

```python
# Listado total de columnas
all_columns_list = data.columns.values.tolist()
all_columns_list
```

```python
# Del total de columnas eliminamos las que contiene desired_columns
sublist = [x for x in all_columns_list if x not in desired_columns]
sublist
```

```python
subset = data[sublist]
subset.head()
```

```python
# Alternativamente:
a = set(desired_columns)
b = set(all_columns_list)
sublist = b - a
sublist = list(sublist)
```

```python
sublist
subset = data[sublist]
subset.head()
```
